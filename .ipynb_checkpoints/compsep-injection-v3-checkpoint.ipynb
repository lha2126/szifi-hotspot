```python
nx = 1024
l = 14.8

n_tiles = 768
nside_tile = 8

### MAKE CUTOUTS
def get_cutout(inp_map, i):
    plt.ioff()
    lon,lat = hp.pix2ang(nside_tile,i,lonlat=True)
    cutout = szifi.get_cutout(inp_map,[lon,lat],nx,l)
    plt.close()
    plt.ion()
    return cutout

def get_tilemap(i):
    """Compute tiling map for a given pixel center"""
    
    smap = np.zeros(hp.nside2npix(nside_tile))
    smap[i] = 1
    umap = hp.ud_grade(smap, 2048)
    return get_cutout(umap, i)
```

```python
def get_mmf_q_map_s(tmap,tem,inv_cov,pix,theta_misc_template=[0.,0.],
mmf_type="standard",cmmf_prec=None,tem_nc=None,comp=0):

    n_freqs = tmap.shape[2]
    y_map = np.zeros((pix.nx,pix.ny))
    norm_map = np.zeros((pix.nx,pix.ny))

    tem_fft = maps.get_fft_f(tem,pix)

    if tem_nc is not None: #Used if cmmf_type == "general"

        tem_nc = maps.get_fft_f(tem_nc,pix)

    filter_fft = get_tem_conv_fft(pix,tem_fft,inv_cov,mmf_type=mmf_type,cmmf_prec=cmmf_prec,
    tem_nc=tem_nc,comp=comp)
    filter = maps.get_ifft_f(filter_fft,pix).real

    tem = maps.get_tmap_times_fvec(tem,cmmf_prec.a_matrix[:,comp]).real #new line

    for i in range(n_freqs):

        y = sg.fftconvolve(tmap[:,:,i],filter[:,:,i],mode='same')*pix.dx*pix.dy
        norm = sg.fftconvolve(tem[:,:,i],filter[:,:,i],mode='same')*pix.dx*pix.dy

        y_map += y
        norm_map += norm

    norm = np.max(norm_map)
    y_map = y_map/norm

    std = 1./np.sqrt(norm)
    q_map = y_map/std

    return q_map,y_map,std

def get_tem_conv_fft(pix_tem,tem_fft,inv_cov,mmf_type="standard",cmmf_prec=None,
tem_nc=None,comp=0):

    if mmf_type == "standard":

        tem_fft = maps.get_tmap_times_fvec(tem_fft,cmmf_prec.a_matrix[:,comp]) #new line
        filter_fft = utils.get_inv_cov_conjugate(tem_fft,inv_cov)

    elif mmf_type == "spectrally_constrained": #doesn't admit changing compoenent order

        if cmmf_prec.cmmf_type == "one_dep":

            n_freq = tem_fft.shape[2]
            a_dot_b =  maps.get_tmap_from_map(cmmf_prec.a_dot_b,n_freq)
            b_dot_b = maps.get_tmap_from_map(cmmf_prec.b_dot_b,n_freq)
            b = cmmf_prec.a_matrix[:,1] #cmmf_prec.sed_b
            a = cmmf_prec.a_matrix[:,0] #cmmf_prec.sed_a

            tem_fft_b = maps.get_tmap_times_fvec(tem_fft,b)
            tem_fft = maps.get_tmap_times_fvec(tem_fft,a)

            filter_fft = utils.get_inv_cov_conjugate(tem_fft,inv_cov) - a_dot_b/b_dot_b*utils.get_inv_cov_conjugate(tem_fft_b,inv_cov)

        elif cmmf_prec.cmmf_type == "general":

            filter_fft_0 = cmmf_prec.w_norm_list[comp]
            filter_fft = filter_fft_0*tem_nc

    filter_fft[np.isnan(filter_fft)] = 0.

    return filter_fft
```

### Create MWE

```python
### Define input spectrum (without noise or beam)

from classy import Class
cosmo = Class()
cosmo.set({'output':'tCl,lCl','lensing':'yes','l_max_scalars':5000})
cosmo.compute()

from scipy.interpolate import interp1d
l_class = cosmo.lensed_cl()['ell']
Cl_class = cosmo.lensed_cl()['tt']*(2.73*1e6)**2

# Interpolate
Cl_interp = interp1d(l_class,Cl_class,bounds_error=False,fill_value=0)

# Noise spectrum
Cl_noise = lambda l: 0.*l

# Setup random fields
import pymaster as nmt
Nx_init = 360 # Initial size (360^2)
dx = 1e-3 # pixel spacing in radians
Nx = 90 # Final size (90^2)

# Model parameters
this_eta_star = 160
this_eta_hs = 280.42513936898837

# GRF function
def create_grf():
    grf = nmt.synfast_flat(Nx_init, Nx_init, Nx_init*dx, Nx_init*dx, [(Cl_interp(l_class)+Cl_noise(l_class))*(l_class>=2)*(l_class<=3500)], [0])[0]
    return grf[:Nx,:Nx]

plt.imshow(create_grf());plt.colorbar();

### Create a signal map
# Load precomputed template and convert to muK
theta_vec = np.linspace(0.,0.1,1000)
model_file = 'model_%s_eta(%.2f,%.2f)_l%d_k(%.2f,%.2f,%d).npy'%('soubhik',this_eta_star,this_eta_hs,3500,0.00,1.00,20)
template_interp = interp1d(theta_vec, np.load(model_file)*(1e6*2.73))

# Define distance from center on pixel grid and interpolate to template
px,py = np.meshgrid(*[np.arange(-Nx//2,Nx//2,1)*dx for _ in range(2)])
dis = np.sqrt(px**2+py**2)
template = template_interp(dis)

# Define the signal map
signal = template.copy()

# Could add a second overlapping signal 14' away with the below!
#cx = 10*np.pi/180./60.
#cy = 10*np.pi/180./60.
#signal += template_interp(np.sqrt((px-cx)**2+(py-cy)**2))

plt.plot(theta_vec,template_interp(theta_vec))
plt.plot(dis[Nx//2],signal[Nx//2],ls='',marker='.')
plt.xlabel(r'$\theta$ [radians]',fontsize=12)
plt.ylabel(r'$\delta T(\theta;\eta_{\star}=160,\eta_{\rm HS}=\eta_{\rm rec})$',fontsize=12);
plt.xscale('log')

plt.figure()
plt.imshow(signal)
plt.colorbar();

### Apply matched filter
# Compute mean power spectrum (this matches CLASS, up to Nx factors)
Pk = np.mean([np.abs(np.fft.fft2(create_grf()))**2 for _ in range(1000)],axis=0)

# Compute FFT of filter
hk = np.fft.fft2(template).conj()

# Compute frequencies
l_vals = np.fft.fftfreq(Nx,d=dx/np.pi/2.)
ll_vals = np.meshgrid(l_vals,l_vals)
modl_vals = np.sqrt(ll_vals[0]**2+ll_vals[1]**2)

# Define l filter
l_filt = (modl_vals>=20)*(modl_vals<=1000)    

# Add a normalization term ( = inverse variance of estimator) 
norm = np.sum(hk*hk.conj()/Pk*l_filt).real
print("Expected noise error: %.2f"%np.sqrt(1./norm))

def compute_Psi(g_value):
    """Compute Psi, normalizing by the template^2 map. This should have an expectation of g"""
    # Define signal
    grf = create_grf()+g_value*signal
    # Compute filter
    return np.fft.ifft2(hk*np.fft.fft2(grf)/Pk*l_filt).real*Nx**2/norm

plt.imshow(np.fft.fftshift(compute_Psi(10)));plt.colorbar()
plt.title(r'$\hat{g} = \Psi/N_\Psi$ Map with $g=10$',fontsize=12);

# ### Histogram max Psi values 
sim_gN = [np.max(compute_Psi(0)) for i in range(500)]
sim_gS_2 = [np.max(compute_Psi(2)) for i in range(500)]
sim_gS_20 = [np.max(compute_Psi(20)) for i in range(500)]

# Alternative approach -- search for signal *at* the injected position
# sim_gN = [compute_Psi(0)[0,0] for i in range(500)]
# sim_gS_2 = [compute_Psi(2)[0,0] for i in range(500)]
# sim_gS_20 = [compute_Psi(20)[0,0] for i in range(500)]

plt.hist(sim_gN,alpha=0.5,label='Noise',bins=30);
plt.hist(sim_gS_2,alpha=0.5,label='Signal+Noise (g=2)',bins=30);
plt.hist(sim_gS_20,alpha=0.5,label='Signal+Noise (g=20)',bins=30);
plt.legend();
plt.xlabel(r'$\hat{g}=\Psi_{\rm max}/N_\Psi$',fontsize=12)
plt.ylabel(r'PDF',fontsize=12);
```

```python
from scipy.interpolate import interp1d

# Gaussian simulations
thetaFWHM = 5.*np.pi/180./60.

from classy import Class
cosmo = Class()
cosmo.set({'output':'tCl,lCl','lensing':'yes','l_max_scalars':5000})
cosmo.compute()

from scipy.interpolate import interp1d
l_class = cosmo.lensed_cl()['ell']
Cl_class = cosmo.lensed_cl()['tt']*pars.TCMB**2*1e12
Cl_interp = interp1d(l_class,Cl_class,bounds_error=False,fill_value=0)

from astropy.io import fits
B_beam = fits.open('/mnt/home/ophilcox/ceph/planck_npipe/COM_CMB_IQU-smica_2048_R3.00_full.fits',field=2)[2].data['INT_BEAM'][:4001]            

beam_func = interp1d(np.arange(len(B_beam)),B_beam,fill_value=0.,bounds_error=False)
beam = beam_func(l_class)
Cl_noise = lambda l: 0.0001*(30.*np.pi/180./60.)**2/(1e-10+beam_func(l))
```

```python
Nx_init = 360
dx = 1e-3
Lx_init = dx*Nx_init
Lx = Lx_init/4
Nx = Nx_init//4

this_eta_star = 160
this_eta_hs = eta_rec

## Create signal map
theta_max = np.max([0.1,np.sqrt(4.*np.pi)*np.asarray(this_eta_star)/eta_0])
theta_vec = np.linspace(0.,theta_max,1000)
model_file = 'model_%s_eta(%.2f,%.2f)_l%d_k(%.2f,%.2f,%d).npy'%('soubhik',this_eta_star,this_eta_hs,3500,0.00,1.00,20)
p_cal_vec = np.load(model_file)*1e6*pars.TCMB
profile_int = interp1d(theta_vec,p_cal_vec, bounds_error=False, fill_value=0.)

px,py = np.meshgrid(*[np.arange(-Lx/2,Lx/2,dx) for _ in range(2)])
dis = np.sqrt(px**2+py**2)
signal = profile_int(dis)
cx = 30*np.pi/180./60.
cy = 40*np.pi/180./60.
signal += profile_int(np.sqrt((px-cx)**2+(py-cy)**2))
```

```python
l_vals = np.fft.fftfreq(len(dis),dx/np.pi/2.)
ll_vals = np.meshgrid(l_vals,l_vals)
modl_vals = np.sqrt(ll_vals[0]**2+ll_vals[1]**2)
l_filt = (modl_vals>=20)*(modl_vals<=1000)

signal_l = np.fft.fft2(signal)
full_filter = signal_l.copy()*l_filt
full_filter[l_filt] /= (Cl_interp(modl_vals)[l_filt]+Cl_noise(modl_vals)[l_filt])

import pymaster as nmt
def create_grf():
    grf = nmt.synfast_flat(Nx_init, Nx_init, Lx_init, Lx_init, [(Cl_class+noise)*(l_class>=2)*(l_class<=3000)], [0])[0]
    return grf[:Nx,:Nx]

def simulator(g_input, return_Phi=False):
    data = create_grf()+g_input*signal
    
    grf_l = np.fft.fft2(data)
    Phi_map = np.fft.ifft2(grf_l*full_filter).real
    if return_Phi:
        return np.fft.fftshift(Phi_map)
    return np.max(data*Phi_map)
```

```python
def load_png_catalog(sim_no, g=0.0):
    i = 0
    for batch in range(77):
        if g==0:
            name = 'outputs_v3/ffp10_sep_%d_batch%d_soubhik.npy'%(sim_no,batch)
        else:
            name = 'outputs_v3/ffp10_sep_%d_batch%d_soubhik_inj-g%.2f.npy'%(sim_no,batch,g)
        try:
            if i==0:
                tot_cat = np.load(name,allow_pickle=True).item()
                i += 1
            else:
                try:
                    tot_cat.append(np.load(name,allow_pickle=True).item())
                except KeyError:
                    print("Key: %s"%batch)
                i += 1
        except IOError:
            #print(batch)
            continue
    if i!=0:
        print("Found %d PNG g=%.2f hotspots (SNR>4) from %d catalogues"%(len(tot_cat.catalogue['q_opt']),g,i))
    return tot_cat

# Cut to SNR of 4
snr_min = 5
sim_no = 0
gs = [0,5,10,20,30,50,100]
png_gs = []
for g in gs:
    png_gs.append(szifi.get_catalogue_q_th(load_png_catalog(sim_no, g=g), snr_min))
```

```python
true_cat = np.load('injection300_v2_sim%d.npz'%sim_no)
all_eta_star = np.geomspace(10,1000,10)

def identify_true_sources(input_cat, dis_arcmin=10):
    
    lon = input_cat.catalogue['lon']
    lat = input_cat.catalogue['lat']

    match_sources = []
    all_close = []
    for t in range(len(true_cat['lon'])):
        
        tdis = hp.rotator.angdist((true_cat['lon'][t],true_cat['lat'][t]),(lon,lat),lonlat=True)
        
        close_ids = np.where(tdis<dis_arcmin*np.pi/180./60.)[0]
        if len(close_ids)==0:
            match_sources.append([np.nan,np.nan,np.nan,np.nan,0])
        else:
            match_sources.append([input_cat.catalogue['lon'][close_ids].mean(),
                                  input_cat.catalogue['lat'][close_ids].mean(),
                                  input_cat.catalogue['q_opt'][close_ids].max(),
                                  input_cat.catalogue['eta_star'][close_ids].mean(),
                                  len(close_ids)])
        all_close.append(close_ids)
            
    return np.asarray(match_sources), np.unique(np.concatenate(all_close))

g_matchs = [identify_true_sources(png_g_proc) for png_g_proc in png_g_procs]
```
