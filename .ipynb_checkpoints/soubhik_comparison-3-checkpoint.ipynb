```python
### Define input spectrum (without noise or beam)
from classy import Class
cosmo = Class()
cosmo.set({'output':'tCl,lCl','lensing':'no','l_max_scalars':5000})
cosmo.compute()

l_class = cosmo.raw_cl()['ell']
Cl_class = cosmo.raw_cl()['tt']*(2.73*1e6)**2

# Interpolate
from scipy.interpolate import interp1d
Cl_interp = interp1d(l_class,Cl_class,bounds_error=False,fill_value=0)

# Define noise spectrum
Cl_noise = lambda l: 0.*l
```

```python
# Setup random fields
Nx_init = 90 # Initial size (360^2)
dx = 1e-3 # pixel spacing in radians
Nx = 90 # Final size (90^2)

# Model parameters
this_eta_star = 160
this_eta_hs = 280.42513936898837

# GRF function
import pymaster as nmt
def create_grf():
    """Create a GRF of size Nx_init x Nx_init (with pixel-width dx) then cut to size Nx x Nx"""
    grf = nmt.synfast_flat(Nx_init, Nx_init, Nx_init*dx, Nx_init*dx, [(Cl_interp(l_class)+Cl_noise(l_class))*(l_class>=2)*(l_class<=3500)], [0])[0]
    return grf[:Nx,:Nx]

plt.title(r'GRF Map [$\mu K$]',fontsize=15)
plt.imshow(create_grf());plt.colorbar();
```

```python
### Create a signal map
# Load precomputed template and convert to muK
theta_vec = np.linspace(0.,0.1,1000)
model_file = 'model_%s_eta(%.2f,%.2f)_l%d_k(%.2f,%.2f,%d).npy'%('soubhik',this_eta_star,this_eta_hs,3500,0.00,1.00,20)
template_interp = interp1d(theta_vec, np.load(model_file)*(1e6*2.73)*(theta_vec<0.020))#5))#025))

# Define distance from center on pixel grid and interpolate to template
px,py = np.meshgrid(*[np.arange(-Nx//2,Nx//2,1)*dx for _ in range(2)])
dis = np.sqrt(px**2+py**2)
template = template_interp(dis)

# # Define the signal map
signal = template.copy()

# # Could add a second overlapping signal 14' away with the below!
# cx = 10*np.pi/180./60.
# cy = 10*np.pi/180./60.
# signal += template_interp(np.sqrt((px-cx)**2+(py-cy)**2))

signalS = maps[101]
template = signal.copy()

# plt.plot(theta_vec,template_interp(theta_vec))
# plt.plot(dis[Nx//2],signal[Nx//2],ls='',marker='.')
# plt.xlabel(r'$\theta$ [radians]',fontsize=12)
# plt.ylabel(r'$\delta T(\theta;\eta_{\star}=160,\eta_{\rm HS}=\eta_{\rm rec})$',fontsize=12);
# plt.xscale('log')

plt.figure()
plt.title(r'Signal [$\mu K$]',fontsize=15)
plt.imshow(signal)
plt.colorbar();
```

```python
ll = np.arange(100,4000,100)
binner = lambda k2: np.asarray([np.sum(k2*(modl_vals<ll[i+1])*(modl_vals>ll[i]))/np.sum((modl_vals<ll[i+1])*(modl_vals>ll[i])) for i in range(len(ll)-1)])

ll_av = 0.5*(ll[1:]+ll[:-1])
```

```python
signal = signalO
template = templateO
```

```python
hkS = np.fft.fft2(templateS).conj()
hkO = np.fft.fft2(templateO).conj()
```

```python
hk_single = np.fft.fft2(template_interp(dis)).conj()
hk_single_cut = np.fft.fft2(template_interp(dis)*(dis<0.02)).conj()
```

```python
plt.plot(ll_av,ll_av**2*binner(hkS*hkS.conj()).real,label='Soubhik')
plt.plot(ll_av,ll_av**2*binner(hkO*hkO.conj()).real,label='Oliver: Pair profile')
plt.plot(ll_av,ll_av**2*binner(hk_single*hk_single.conj()).real,label='Oliver: Single profile')
plt.plot(ll_av,ll_av**2*binner(hk_single_cut*hk_single_cut.conj()).real,label=r'Oliver: Single profile, $\theta$ < 0.02')
plt.yscale('log');
plt.legend(fontsize=12);
```

```python
### Apply matched filter
# Compute mean power spectrum (this matches CLASS, up to Nx factors)
Pk = np.mean([np.abs(np.fft.fft2(create_grf()))**2 for _ in range(1000)],axis=0)

# Compute FFT of filter
hk = np.fft.fft2(template).conj()

# Compute frequencies
l_vals = np.fft.fftfreq(Nx,d=dx/np.pi/2.)
ll_vals = np.meshgrid(l_vals,l_vals)
modl_vals = np.sqrt(ll_vals[0]**2+ll_vals[1]**2)

# Define l filter
l_filt = (modl_vals>=20)*(modl_vals<=1000)   

# Add a normalization term ( = inverse variance of estimator) 
norm = np.sum(hk*hk.conj()/Pk*l_filt).real
print("Expected noise error: %.2f"%np.sqrt(1./norm))

def compute_Psi(g_value):
    """Compute Psi, normalizing by the template^2 map. This should have an expectation of g"""
    # Define signal
    grf = create_grf()+g_value*signal
    # Compute filter
    return np.fft.ifft2(hk*np.fft.fft2(grf)/Pk*l_filt).real*Nx**2/norm

plt.imshow(np.fft.fftshift(compute_Psi(10)));plt.colorbar()
plt.title(r'$\hat{g} = \Psi/N_\Psi$ Map with $g=10$',fontsize=12);
```

```python
### Histogram max Psi values 
sim_gN = [np.max(compute_Psi(0)) for i in range(500)]
sim_gS_2 = [np.max(compute_Psi(2)) for i in range(500)]
sim_gS_20 = [np.max(compute_Psi(20)) for i in range(500)]

# # Alternative approach -- search for signal *at* the injected position
# sim_gN = [compute_Psi(0)[0,0] for i in range(500)]
# sim_gS_2 = [compute_Psi(2)[0,0] for i in range(500)]
# sim_gS_20 = [compute_Psi(20)[0,0] for i in range(500)]
```
